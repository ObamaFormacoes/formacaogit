```
arrayPal = ["arara", "biscoito", "ata", "asa"]

arrayRes = [True, False, True, True]
```

```
def isPal(array):
  result = []
  for word in array: 
    if (''.join(reversed(word)) == word):
      result.append(True)
    else:
      result.append(False)
  return result
```

```
print(isPal(arrayPal))
```

```
def compare(entry, correct):
  if (entry == correct):
    print("Palavras palíndromas detectadas corretamente!")
  else:
    print("Palavras palíndromas detectadas de forma errônea!")
```

```
compare(isPal(arrayPal),arrayRes)
```
